```python
dfv["tags"]
```

```python
def leng(val):
    if val == "[none]":
        return 0
    
    return len(val.split("|"))

dfv["tag_conteo"] = dfv["tags"].apply(leng)

dfv
```

```python
#parte 5
dfv["Likes"] = dfv["likes"] > dfv["dislikes"]
#dfv[dfv["Likes"]==False]
dfv.head()
```

```python
#parte 6  
import json

with open("./data/US_category_id.json") as data_file:    
    data = json.load(data_file)  

dfCategory = pd.json_normalize(data, 'items', record_prefix='category_')
dfCategory.rename(columns={'category_snippet.title': 'category_title', 
                           'category_snippet.channelId':'category_channelId',
                           'category_snippet.assignable':'category_assignable'}, 
                inplace=True)
dfCategory.head()
```

```python
dfv.head()
```

```python
#Parte 7
dfCategory["category_id"]=dfCategory["category_id"].astype(int)

dfVideoCat = pd.merge(dfv, dfCategory,on="category_id")

dfVideoCat = dfVideoCat[["title","channel_title","category_title","views","likes","dislikes","comment_count","description","tag_conteo","Likes"]]
dfVideoCat.head()

```

```python
#Partes 8
dfTop20 = dfVideoCat.sort_values(by="views",ascending=False)[:20]
dfTop20["rank"] = np.arange(len(dfTop20)) 
dfTop20["rank"] =  dfTop20["rank"] + 1
#echo(dfTop20[["rank","views","title"]])

for r,v,t in zip(dfTop20["rank"],dfTop20["views"],dfTop20["title"]):
    x = str(r).strip()+": views:"+str(v).strip()+" title:"+t.strip()
    print(x)
    print('-----------------------------------------------------------------------------------------')
```
