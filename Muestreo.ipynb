```python
import numpy as np
import scipy as sp
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd 
```

```python

Dist_Binomial = stats.binom(25,.7)

dado = stats.randint(1,7)

exp = stats.expon(0,10)


fig, axs = plt.subplots(1,3,figsize=(12,4))

axs[0].plot(range(0,25),Dist_Binomial.pmf(range(0,25)))
axs[1].plot(range(1,7),dado.pmf(range(1,7)))
axs[2].plot(range(1,15),exp.pdf(range(1,15)))



fig.tight_layout()
```

```python
np.random.seed(100)

binom_sample = Dist_Binomial.rvs(200);
binom_sample
```

```python
np.random.seed(100)
unif_sample= dado.rvs(200)*100
unif_sample
```

```python
np.random.seed(100)
exp_sample = exp.rvs(200) 
exp_sample
```

```python
def boostrap(data,n_samples=100):
    boostrap_medians = []
    boostrap_mean =[]
    boostrap_23=[]
    
    for i in range(n_samples):
        boostrap_sample = np.random.choice(data,size=len(data),replace=True)
        boostrap_medians.append(np.median(boostrap_sample))
        boostrap_mean.append(np.mean(boostrap_sample))
        boostrap_23.append(np.percentile(boostrap_sample,23))

    fig,ax = plt.subplots(1,figsize=(15,4))
    ax.hist(boostrap_medians,bins=25,density=True,color="black",alpha=.4,label="Medianas")
    ax.hist(boostrap_mean,bins=25,density=True,color="red",alpha=.75,label="Medias")
    ax.hist(boostrap_23,bins=25,density=True,color="blue",alpha=.55,label="Medias")
    
    return boostrap_medians,boostrap_mean,boostrap_23
    
    
```

```python
binom_median,binom_mean,binom_23=boostrap(binom_sample)
```

```python
exp_median,exp_mean,exp_23=boostrap(exp_sample)
```

```python
unif_median,unif_mean,unif_23=boostrap(unif_sample)
```

```python
doc = open("./data/productivity.txt","r")

sample = doc.readlines()

sample_values = np.array(sample).astype(np.float)

sample_values
```

```python
def  boostrap_ci(samples_values,stat_function = np.mean,resamples = 10000,ci=95):
    statistics = np.zeros(resamples)
    for i in range(resamples):
        boostrap=np.random.choice(samples_values,len(sample_values),replace=True)
        statistic = stat_function(boostrap)
        statistics[i]=statistic
    lower_bound = (100-ci)/2
    return np.percentile(statistics,lower_bound),stat_function(samples_values),np.percentile(statistics,100-lower_bound),statistics

    
    
    
    
    
    
```

```python
lower,estimate,upper,boostrap=boostrap_ci(sample_values)

print(lower,estimate,upper);

plt.hist(boostrap);
```
