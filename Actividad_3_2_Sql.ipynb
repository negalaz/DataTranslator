```python
import numpy as np
import pandas as pd
import sqlite3 # <-- Interactuar con BD SQLite
import matplotlib.pyplot as plt
%matplotlib inline

conn = sqlite3.connect("data/hr.sqlite"); conn
```

```python
#Mostrar los nombres (first_name, last_name) utilizando el alias “first name”, “last name” de la tabla de empleados
sql = """
select first_name, last_name
from employees
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener el “departmet ID” único de la tabla de empleados
pd.read_sql_query("""
    SELECT department_id
    FROM employees
    GROUP BY department_id
""",conn)
```

```python
#Obtener todos los detalles de los empleados de la tabla de empleados ordenados por nombre, de manera descendente
pd.read_sql_query("""
    SELECT *
    FROM employees
    ORDER BY first_name desc
""",conn)
```

```python
#Obtener nombre, apellido, salario y la remuneración de todos los empleados (la remuneración se calcula como el 12% del salario).
pd.read_sql_query("""
    SELECT first_name
        ,last_name
        ,salary
        ,(salary * .12) as remuneracion
    FROM employees 
""",conn)
```

```python
#Obtener el  “employee ID”, nombre, apellido y salario ordenados de manera ascendente por el salario.
pd.read_sql_query("""
    SELECT employee_id
        ,last_name
        ,salary
    FROM employees 
    ORDER BY salary DESC
""",conn)
```

```python
#Obtener el total del salario pagado a los empleados.
pd.read_sql_query("""
    SELECT sum(salary) as TotalPagado
    FROM employees 
    ORDER BY salary DESC
""",conn)
```

```python
#Obtener el salario máximo y mínimo de la tabla de empleados.

pd.read_sql_query("""
    SELECT 'Salario Maximo' as Descripcion, MAX(salary)
    FROM employees 
    UNION ALL
    SELECT 'Salario Minimo' as Descripcion, MIN(salary)
    FROM employees 
    
""",conn)
```

```python
#Obtener el salario medio y el número de empleados de la tabla de empleados.
sql = """
select avg(salary) [Salario Promedio]
, count(employee_id) [Numero de Empleados]
from   employees
"""
pd.read_sql_query(sql, conn)
```

```python
pd.read_sql_query("""
    select count(distinct job_id) as Puestos 
    from employees
    """,conn)
```

```python
#Obtener todos los nombres de la tabla de empleados en mayúsculas.
pd.read_sql_query("""
    select upper(first_name) as nombre
    from   employees
    """,conn)
```

```python
#Obtener los 3 primeros caracteres del nombre de cada empleado.
pd.read_sql_query("""
select substr(first_name, 1, 3) 
from employees""", conn)
```

```python
#Obtener el nombre completo de todos los empleados de la tabla de empleados.
pd.read_sql_query("""
select first_name || ' ' || last_name as [Nombre Completo]
from employees""", conn)
```

```python
#Obtener el salario mensual (a 2 decimales) de todos los empleados.
pd.read_sql_query("""
select round(salary,2) as salario 
from employees""", conn)
```

```python
#Obtener nombre, apellido y salario de todos los empleados cuyo salario no esté en el rango de 10,000 a 15,000.
pd.read_sql_query("""
select first_name, last_name, salary
from   employees
where  salary < 10000 or salary > 15000
""", conn)
```

```python
#Obtener nombre, apellido y departamento de todos los empleados de los departamentos 30 o 100 en orden alfabético por “department ID”.
pd.read_sql_query("""
select e.first_name, e.last_name, department_name 
from employees e 
join department d 
on e.department_id = d.department_id 
where e.department_id in (30, 100) 
order by department_name asc
""", conn)
```

```python
#Obtener nombre, apellido y salario de todos los empleados cuyo salario 
#no esté en el rango de 10,000 a 15,000 y estén en el departamento 30 o 100.
pd.read_sql_query("""
select first_name
, last_name
, salary
from   employees
where  salary < 10000 or salary > 15000
    and department_id in (30,100)
""", conn)
```

```python
#Obtener el nombre de todos los empleados que tengan tanto una "b" y "c" en su nombre.
pd.read_sql_query("""
select first_name 
from employees 
where first_name like '%b%c%' 
""", conn)
```

```python
#Obtener el apellido, el cargo y el salario de todos los empleados 
#cuyo trabajo es el de “Programmer” o “Shipping Clerk”, y cuyo salario no sea igual a 4,500, 10,000 o 15,000 dólares.
pd.read_sql_query("""
select e.last_name
, j.job_title
, e.salary
from   employees e
join   jobs j
   on  e.job_id = j.job_id
where  e.salary not in (4500, 10000, 15000)
   and j.job_title in ('Programmer', 'Shipping Clerk')
""", conn)
```

```python
#Obtener los apellidos de los empleados cuyos nombres tienen exactamente 6 caracteres.
pd.read_sql_query("""
select last_name 
from employees 
where length(first_name) = 6
""", conn)
```

```python
#Obtener los apellidos de los empleados que tienen "e" como tercer carácter.
pd.read_sql_query("""
select e.last_name
from   employees e
where  substr(last_name, 3, 1) = 'e'
""", conn)
```

```python
#Obtener el salario máximo de un empleado que trabaja como “Programmer”.
pd.read_sql_query("""
select max(e.salary) [Salario Maximo Programador]
from   employees e 
join   jobs j 
on  e.job_id = j.job_id
where  j.job_title = 'Programmer'
""", conn)
```

```python
#Obtener el salario medio y el número de empleados que trabajan en el departamento 90.
pd.read_sql_query("""
select avg(e.salary) as [Salario Promedio]
    , count(employee_id) [Empleados]
from   employees e
where  e.department_id = 90
""", conn)
```

```python
#Obtener el salario más alto, el más bajo, la suma y la media de todos los salarios.
pd.read_sql_query("""
select max(salary) as max
, min(salary) as min
, sum(salary) as sum
, avg(salary) as avg 
from employees
""", conn)
```

```python
#Obtener el número de empleados con el mismo puesto.
pd.read_sql_query("""
select j.job_title as [puesto]
, count(e.employee_id) as [numero de empleados]
from   employees e
join   jobs j
   on  e.job_id = j.job_id
group by j.job_title
""", conn)
```

```python
#Obtener la diferencia entre el salario más alto y el más bajo.
pd.read_sql_query("""
select max(salary) as max
, min(salary) as min
, max(salary) - min(salary) as diferencia
from employees 
""", conn)
```

```python
#Obtener el “manager ID” y el salario del empleado con menor sueldo reportando a ese “manager''.
pd.read_sql_query("""
select e.manager_id
, min(salary) [Salario minimo]
from   employees e
group by e.manager_id
""", conn)
```

```python
#Obtener el “department ID” y el total de salarios a pagar en cada departamento.
pd.read_sql_query("""
select e.department_id
, sum(salary) as [total]
, count(salary) as [cantidad] 
from   employees e 
group by e.department_id
""", conn)
```

```python
#Obtener el salario medio de cada ID de puesto, excluyendo “Programmer”.
pd.read_sql_query("""
select j.job_id
, avg(e.salary) 
from   employees e
join   jobs j
   on  e.job_id = j.job_id
where  j.job_title <> 'Programmer'
group by j.job_id
""", conn)
```

```python
#Obtener el job ID y el salario máximo de los empleados que su salario máximo es mayor o igual a $4,000.
pd.read_sql_query("""
select e.job_id, max(salary) max_salary
from   employees e 
group by e.job_id 
having max(salary) >= 4000
""", conn)
```

```python
#Obtener el salario medio de todos los departamentos que emplean a más de 10 trabajadores.
pd.read_sql_query("""
select d.depart_name
, avg(e.salary) 
from   employees e
join   departments d
   on  e.department_id = d.department_id
group by d.depart_name
having count(employee_id) > 10
""", conn)
```

```python
#Obtener nombre, apellido y  salario de los empleados que tienen un salario más alto 
#que el empleado cuyo apellido es “Bull”.
pd.read_sql_query("""
select e.first_name
, e.last_name
, e.salary 
from   employees e 
where  salary > (select e2.salary 
                 from employees e2 
                 where last_name = 'Bull') 
""", conn)
```

```python
#Obtener nombre y apellido de todos los empleados que trabajan en el departamento de IT.
pd.read_sql_query("""
select e.first_name
     , e.last_name
from   employees e
join   departments d
   on  e.department_id = d.department_id
where  d.depart_name = 'IT'
""", conn)
```

```python
#Obtener nombre y apellido de los empleados que tienen un “manager” que trabaja en un departamento con sede en EUA.
pd.read_sql_query("""
select e.first_name
, e.last_name
, m.first_name 
from   employees e 
join employees m 
    on e.manager_id = m.employee_id
join departments d 
    on m.department_id = d.department_id
join locations l 
    on d.location_id = l.location_id
join countries c 
    on l.country_id = c.country_id 
where c.country_id = 'US'
""", conn)
```

```python
#Obtener nombre y apellido de los empleados que son “managers”.
pd.read_sql_query("""
select distinct m.first_name, m.last_name
from   employees e
join   employees m
  on e.manager_id = m.employee_id
""", conn)
```

```python
pd.read_sql_query("""
select e.first_name
, e.last_name
, e.salary 
from   employees e 
where  salary > (select avg(em.salary) 
                 from employees em) 
""", conn)
```

```python
#Obtener nombre, apellido y salario de los empleados cuyo salario es igual al salario mínimo para su rol.
pd.read_sql_query("""
select e.first_name, e.last_name
from   employees e
join   jobs j
    on e.job_id = j.job_id
    and e.salary = j.min_salary
""", conn)
```

```python
#Obtener nombre, apellido y salario de los empleados que ganan más que el 
#salario medio y que trabajan en cualquier área de IT.
pd.read_sql_query("""
select e.first_name
    , e.last_name
    , e.salary
    , department_name 
from   employees e 
join department d 
 on e.department_id = d.department_id 
where  salary > (select avg(em.salary) from employees em)
and d.department_name like 'IT%'
""", conn)
```

```python
#Obtener nombre, apellido y salario de los empleados que ganan más que el Sr. Bell.
pd.read_sql_query("""
select e.first_name
    , e.last_name
    , e.salary
from   employees e
where  salary > (select em.salary 
                 from employees em 
                 where last_name = 'Bell')
""", conn)
```

```python
#Obtener nombre y apellido de los empleados que no son “supervisors”.
pd.read_sql_query("""
select em.first_name, em.last_name
from   employees em
where  em.employee_id not in
                (   select distinct m.employee_id
                    from   employees e
                    join   employees m
                      on e.manager_id = m.employee_id
                )
""", conn)
```

```python
#Obtener el employee ID, nombre, apellido y salario de todos los empleados 
#cuyo salario es superior a la media de sus departamentos.
pd.read_sql_query("""
select em.first_name
, em.last_name
, em.last_name
from   employees em
join (
    select d.department_id, avg(e.salary) as avg_salary
    from   employees e
    join   departments d
    on  e.department_id = d.department_id
    group by d.depart_name
) dep
    on  em.department_id = dep.department_id
where   em.salary > dep.Avg_Salary
""", conn)
```

```python
#Obtener el 5º salario máximo de la tabla de empleados.
pd.read_sql_query("""
select first_name
, last_name
, salary 
from (select e.first_name, e.last_name, e.salary 
      from   employees e 
      order by salary desc 
      Limit 5) 
order by salary asc limit 1
""", conn)
```

```python
#Obtener las direcciones (location_id, street_address, city, state_province, country_name) de todos los departamentos.
pd.read_sql_query("""
select d.depart_name
, d.location_id
, l.street_address
, l.city
, state_province
, c.country_name 
from departments d 
join locations l 
    on d.location_id = l.location_id 
join countries c 
    on l.country_id = c.country_id
where c.country_id = 'US'
""", conn)
```

```python
#Obtener el “employee ID”, nombre y apellido de cada empleado junto con el nombre, apellido y “manager ID” de su “manager”.
pd.read_sql_query("""
select e.employee_id
, e.first_name
, e.last_name
, m.first_name as [nombre manager]
, m.last_name as [apellido manager]
from   employees e
join   employees m
  on e.manager_id = m.employee_id
""", conn)
```

```python
#Obtener el job_title y el salario medio de los empleados.
pd.read_sql_query("""
select j.job_title
, avg(e.salary) 
from   employees e
join   jobs j
   on  e.job_id = j.job_id
group by j.job_title
""", conn)
```
